```python
#se utiliza la funcion plot de la libreria matplot para graficar una funcion"y" definida
x = np.linspace(-3,3,50)
y=x**2
plt.grid(True)
plt.plot(x,y,"r")
```

```python
#se grafica una funcion seno/cos  que va desde -pi a pi.
x = np.linspace(-np.pi,np.pi,100)
y1= np.cos(x)
y2= np.sin(x)
plt.grid(True)
plt.plot(x,y1,x,y2,"r")
```
