#### Niveles  de autonomia

La conduccin autnoma ha sido clasificada en seis niveles por la Sociedad de Ingenieros de Automocin (SAE, por sus siglas en ingls). La actualizacin ms reciente es el estndar J3016.

### No hay automatizacin( Nivel 0):
El nivel 0 de los vehculos autnomos que aplica a los coches ms comunes en la actualidad, no se automatizan las acciones del conductor.

###### Caractersticas:

El conductor realiza todas las tareas y maniobras.

##### Limitaciones:

### Conduccin asistida( Nivel 1):
Los vehculos autnomos de nivel 1 incluyen sistemas que controlan la direccin, la velocidad y el frenado. Muchos modelos disponibles en la actualidad incluyen funciones de este nivel, entre otras:

###### Caractersticas:

Se mantienen las manos en el volante.

##### Limitaciones:
